```python
trait_name = "attractive"
meanRT = {}
sub_folder_prefix = ''
df = pd.read_csv(f"../{trait_name}/likert_data.csv")
s = sorted(df.subId.unique())
subs = {sub: {"meanRT": None, "stdRT": None, "selfCorr": None, "GroupCorr": None, "RatingSTD": None} for sub in s}


# Mean rt: a good cut off is 2 seconds
for sub in s:
     subs[sub]["meanRT"] = np.mean(df[df.subId == sub].rt)
        
# std rt
for sub in s:
     subs[sub]["stdRT"] = np.std(df[df.subId == sub].rt)
        
# Self Consistency
selfConsistency = calcSelfConsistency(trait_name)
for i, sub in enumerate(s):
    subs[sub]["SelfCorr"] = selfConsistency[i]
    
# Group Consistency
groupConsistency = calcGroupConsistency(trait_name)
for i, sub in enumerate(s):
    subs[sub]["GroupCorr"] = groupConsistency[i]
    
# RatingSTD
for sub in s:
    subs[sub]["RatingSTD"] = np.std(df[df.subId == sub].rating)
```
